```python
import pandas as pd, numpy as np, seaborn as sns
```

```python
wt = pd.read_csv('./data/weather.csv', engine='python')
```

```python
wt['일시'].str[-2:]
```

```python
wt['MONTH'] = wt['일시'].str[0:3]
wt['YEAR'] = '20' + wt['일시'].str[-2:]
```

```python
import calendar

d = {v: str(k).rjust(2, '0') for k,v in enumerate(calendar.month_abbr)}
```

```python
wt['MONTH'] = wt['MONTH'].map(d)
```

```python
wt['YEAR']
```

```python
wt['DATE'] = wt['YEAR'] + wt['MONTH']
```

```python
wt.drop(['일시','지점','YEAR','MONTH'], axis=1, inplace=True)
```

```python
g_wt = wt.groupby(['DATE','시도']).mean().reset_index()
```

```python
new_wt = g_wt[g_wt['DATE']!='202007']
```
