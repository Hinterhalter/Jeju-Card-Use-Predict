```python
import pandas as pd, numpy as np, seaborn as sns
```

```python
df = pd.read_csv('./data/final_df.csv')
wt = pd.read_csv('./data/new_weather.csv')
```

```python
df.head()
```

```python
wt.head()
```

```python
wt.rename(columns={'SIDO':'CARD_SIDO_NM',
                   'DATE':'REG_YYMM'}, inplace=True)
```

```python
total_df = pd.merge(df, wt, on=['CARD_SIDO_NM','REG_YYMM'])
```

```python
total_df.columns
```

```python
sns.heatmap(total_df[['AGE', 'SEX_CTGO_CD',
       'FLC', 'CSTMR_CNT', 'AMT', 'CNT', 'CV', 'AVG_TEMP',
       'AVG_MAX_TEMP', 'AVG_MIN_TEMP', 'MONTH_PRECIPITATION']].corr(),cmap='YlGnBu',fmt='.1g')
```
